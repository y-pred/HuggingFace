```python
%%writefile app.py
import streamlit as st
from transformers import pipeline

#Python function for summarization
def generate_summary(text,max_len):
  summarizer = pipeline('summarization')
  return summarizer(text,max_length=max_len)[0]['summary_text']

#Streamlit app
def main():
  st.title('Text Summarization')

  #Text input
  text_input = st.text_area('Enter your text here')

  #Slider for number of words
  num_words = st.slider('Number of words in summary', min_value=10, max_value=100, value=50)

  #Button to trigger summarization
  if st.button('Summarize'):
    if text_input:
      #Generate summary
      summary = generate_summary(text_input,num_words)

      #Display summary
      st.subheader('Summary:')
      st.write(summary)

    else:
      st.error('Please enter some text')


if __name__ == "__main__":
  main()


```
